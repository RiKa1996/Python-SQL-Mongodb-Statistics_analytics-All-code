## **Event Handling**
- isme agar hum Entry me kuch type kare to to interect hona chaiye
- any user interaction with any graphical is known as event
- Handling of event means reponding to that event i.e. execution of a logic.

## **Step in Event Handling**
- Identify source & type of event
- define a function that contains logic (a.k.a handler)
- bind source with handler

## **Source**
- any graphical component (win,button,label,etc.) can be a source

# **Type**
- Enter
- Leave
- Button
- Key
- etc.

```python
from tkinter import *
from tkinter import messagebox
win=Tk()
win.geometry("600x400")
win.configure(bg="blue")

def change1(event):
    b1.configure(fg="green")
    
def change2(event):
    b1.configure(fg="black")
    
def change3(event):
    b2.configure(fg="blue")
    
def change4(event):
    b2.configure(fg="black")

#eval function    
def result():
    a=eval(e2.get())
    messagebox.showinfo("Result",f"Answer:{a}")
    
def reset():
    e2.delete(0,"end")
    
    
l1=Label(win,text="Basic Calculator",font=("",30,"bold"),bg="blue",fg="black")
l1.pack()

l2=Label(win,text="Enter Expression:",font=("",15,'bold'),bg="blue",fg="black")
l2.place(x=100,y=100)

e2=Entry(win,font=("",15,"bold"),bd=5)
e2.place(x=280,y=100)

#eval function
b1=Button(win,text="Result",font=("",15,"bold"),width=6,command=result)
b1.place(x=275,y=250)
b1.bind("<Enter>",change1)                                     #handler
b1.bind("<Leave>",change2)

#eval function
b2=Button(win,text="Reset",font=("",15,"bold"),width=6,command=reset)
b2.place(x=375,y=250)
b2.bind("<Enter>",change3)                                     #handler
b2.bind("<Leave>",change4)


win.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog
win=Tk()
win.geometry("600x400")
win.configure(bg="blue")

def change1(event):
    b1.configure(fg="green")
    
def change2(event):
    b1.configure(fg="black")
    
def change3(event):
    b2.configure(fg="blue")
    
def change4(event):
    b2.configure(fg="black")

#eval function    
def result():
    s=e2.get()
    if(s=='' or len(s)==1):
        messagebox.showerror("validation","please write expression")
        e2.focus()
        return
    a=eval(s)
    messagebox.showinfo("Result",f"Answer:{a}")
    
def browse():
    filepath=filedialog.askopenfilename()
    file=open(filepath,'r')
    text=file.read()
    file.close()
    e2.delete(0,"end")                    #ye phle se expression ko delete kr deta hai.
    e2.insert(0,text)                     #ye expression ko add krta hai.
    
    
    
    
l1=Label(win,text="Basic Calculator",font=("",30,"bold"),bg="blue",fg="black")
l1.pack()

l2=Label(win,text="Enter Expression:",font=("",15,'bold'),bg="blue",fg="black")
l2.place(x=100,y=100)

e2=Entry(win,font=("",15,"bold"),bd=5)
e2.place(x=280,y=100)

#eval function
b1=Button(win,text="Result",font=("",15,"bold"),width=6,command=result)
b1.place(x=275,y=250)
b1.bind("<Enter>",change1)                                     #handler
b1.bind("<Leave>",change2)

#eval function
b2=Button(win,text="browse",font=("",15,"bold"),width=6,command=browse)
b2.place(x=375,y=250)
b2.bind("<Enter>",change3)                                     #handler
b2.bind("<Leave>",change4)


win.mainloop()
```
