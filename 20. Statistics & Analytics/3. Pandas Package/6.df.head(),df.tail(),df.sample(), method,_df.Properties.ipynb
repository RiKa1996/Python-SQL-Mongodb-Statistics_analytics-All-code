- **df.sample method**

- **Columns ke Name change krna**
- **df.rename method**
    - rename method se row ke index aur columns ke lables ko change kr skte hai.
    - **numpy me**
        - sum(a,axis=0)----->column
        - mean(a,axis=1)---->row
        
    - **pandas me**
        - axis=0------->row
        - axis=1------->column

```python
#rename method se row ke index aur columns ke lables ko change kr skte hai.
#abhi humne columns ke lables ko change kiya hai.
df.rename({'total_bill':'Bill','tip':'Tip'},axis=1)
```

```python
#row ke index ko change krege.
#inplace=True argument likhne per exiting table fixed ho jati hai.---------aur aage hum is table ka use krege tab same data hume milega.
df.rename({1:500,2:'india',3:'rishav'},axis=0,inplace=True)    #change in existing dataframe
df
```
