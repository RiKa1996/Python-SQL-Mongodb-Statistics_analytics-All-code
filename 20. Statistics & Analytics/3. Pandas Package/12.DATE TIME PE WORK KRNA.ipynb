- **important**

```python
df['date_time']=pd.to_datetime(df.Time)
```

### **dt properties**
- ye datetime se related hota hai.
- ye date_time column me date, time, year, day, ki series deta hai.

```python
df["year"]=df.date_time.dt.year
df["month"]=df.date_time.dt.month_name()
df["day"]=df.date_time.dt.day_name()
```

```python
df
```

```python
df.year.value_counts()
```

```python
df.day.value_counts()
```

```python
df.month.value_counts()
```

```python
#hour,minute,second ko ek data se nikalna
df["hour"]=df.date_time.dt.hour
```
