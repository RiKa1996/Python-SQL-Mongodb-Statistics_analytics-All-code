### **loc and iloc** se hi index aur column ko slicing kr skte hai.
- **df.loc**[startIndex:stopIndex:step,startLabel:stopLabel:setp]
- loc and iloc se index aur column slicing krte hai to-----------**yha start and stop dono hi inclusive hota hai.**
- **And**
- **df.iloc**[startIndex:stopIndex:step,startLabel:stopLabel:setp]
- loc and iloc se index aur column slicing krte hai to-----------**start------inclusive** and **stop------exclusive**
- ydi iloc slicing me stop bata de to **exclusive** hoga.
- iloc me **stop** inclusive hota hai----jb by default slicing me stop nhi batate hai.------>**df.iloc[0::1,0::]**

```python
df
```

```python
#df.loc[row,column]----------->df.loc[startIndex:stopIndex:step,startLabel:stopLabel:setp]
df.loc['india':3:1,'total_bill':'day':1]
```

```python
df.loc[5:50:5,'total_bill':'size':2]
```

```python
#iloc ke sath integer index and column index , slicing
df.iloc[2:6:1,0:7:1]                     #start---inclusive        stop----exclusive
```

```python
df.iloc[0::1,0::]                         #step---default step 1 ------------stop-default stop---inclusive
```

```python
df.iloc[5::,:3]
```

```python
df.iloc[5:,::-1]
```
