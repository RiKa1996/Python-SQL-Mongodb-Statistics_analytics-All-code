### **database se pandas me data ko lana** then pandas se us database ko **excel and csv file me convert krna**
- form of URL=**"dialect[+driver]://user:password@host/dbname[?key=value..]"**
    - dialect=database ka name------like: mysql
    - driver=**pymysql** module
    

```python
#form of URL=``dialect[+driver]://user:password@host/dbname[?key=value..]``
eng=sqlalchemy.create_engine("mysql+pymysql://root:root@localhost/11to1230")
df=pd.read_sql("select * from emp",eng)
df
```

```python
#to_excel------->pandas se excel file me data store krna
df.to_excel("E:/New folder (2)/20. Statistics & Analytics/3. Pandas/stu(to_excel).xlsx",index=False)
```

```python
#to_csv---------->pandas se csv file me data ko store krna.
df.to_csv("E:/New folder (2)/20. Statistics & Analytics/3. Pandas/stu(to_csv).csv",index=False)
```
