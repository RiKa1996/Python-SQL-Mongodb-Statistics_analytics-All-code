- **nrows** number of rows ------>jaise koi data jyada bada, but hum us data se kuch rows padna chahte hai to nrows argument ka use kr skte ha.
- **skiprows** -------jo number of rows hum likh rhe , ye excel jo store ki hai humne usme se skip kr ke ,,no. of rows likhne deta hai.
    - skiprows me alternate ya range function se bhi rows skip ho jata hai ,ek list bana kr jaise:- skiprows=[1,2,5,63,55,] ye skiprows=[1:15:2]
    - skiprows me range function se bhi skip kr skte hai.------**skiprows=range(1,5)**

```python
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=5)     #excel data ke suru ke 5 rows skip kr ke 6th row ko header bana deta hai.
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=[0,2,3,5,6,22,20])  #excel data me alternate skip krne,ydi 0 index row ko skip krte hai 1sr row ko header bana dega
df
```

```python
#range function se skip krna;
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=range(1,6,2))
df
```

- **index_col=[]** argument-------for example Title,Year,Language columns ko hum ydi column index banana chahte hai to .
- **usecols=['Title','Year','Language']** argument

```python
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=range(1,6,2),usecols=['Title','Year','Language'])
df
```

```python
#(e.g. "A:E" or "A,C,E:F")----aise bhi likh skte hai.
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=range(1,6,2),usecols="A,B,F")  #usecols me excel ke default column ke name likhte hai. like:- "A,B,C"
df
```

```python
df=pd.read_excel("E:/dataset/analysis/movies_sheets.xls",nrows=100,skiprows=range(1,6,2),usecols=['Title','Year','Language'],index_col=[1])
df
```
