- **1. fillna() method**
    - fillna=fill not available.
    - ye method toda galat ho jata hai kyoki  ye har NaN wali jagaha pe same value putt ho jata hai. jiske ki pura data galat ho jata hai.

```python
#missing values ko fill krege.
#2 way to handle missing values-------->1.fill the missing value-------------2.delete row/columns containg missing vlaue
df.fillna(20)          #returns a new dataframe after filling 20
```

```python
df
```

```python
#is bari alag alag columns se missing values put krege.----by dist.
df.fillna({'temperature':30,'windspeed':10,'event':"unknow"})
```

- **2. ffill() method**
    - ffill---forward filling
    - is method ka drawback-----ydi phli row me koi values missing hai to , ye method ruk skta hai.

```python
df.ffill()
```

```python
#ffill method se kisi perticular coloumn pe lagana chahte hai to.
df.temperature.ffill()
```

```python
df['temperature']=df.temperature.ffill()
df['windspeed']=df.windspeed.bfill()
```

```python
df
```

    - bfill===back filling===back se data ko fill krna.

```python
df.at[3,'event']="Sunny"     #at properties se ydi koi values last me, suruaat me , ya kai beech me missing ho to phle at properties se fill kr skte hai then bfill method use kr skte hai.
df
```

### **1.1 handling missing values**

```python
df=pd.read_csv("e:/dataset/analysis/weather_data.txt")
df
```

- **4. dropna() method**
    - dropna=drop not available
    - isme row / columns puri ki puri drop ho jayega

```python
df.dropna(axis=0)     #drop rows
```

```python
df.dropna(axis=1)       #drop cols   #columns wise data ko nhi drop krna chaiye
```

```python
df
```

```python
#thresh: 'int | NoDefault' = <no_default>,------thresh=3 krne per jis row me 3 se kam values hai wo delete ho jayegi.
df.dropna(axis=0,thresh=3)                    #if row contain<3 non-missing values then delete it
```

```python
#thresh: 'int | NoDefault' = <no_default>,------thresh=3 krne per jis row me 3 se kam values hai wo delete ho jayegi.
df.dropna(axis=0,thresh=2)                    #if row contain<3 non-missing values then delete it
```

```python
#specific column ko bacha ke alag column delete kr skte hai.-------ye tecnique sbse jyada use hota hai. industry me.
#subset[""]
df.dropna(axis=0,subset=["temperature"])
```

```python
df.dropna(axis=0,subset=["temperature",'windspeed'])
```

```python
df
```

### **2. handling duplicacy**


```python
df=pd.read_csv("e:/dataset/analysis/weather_data2.txt")
df
```

- **6. drop_duplicates() method**
    - dublicates pahchanne ka paimana--->har row ki values same honi chaiye dusri row ki same position ki value se.

```python
df.drop_duplicates()
```

```python
#apne hisab se row ke jariye duplicasy pahchanna.
df.drop_duplicates(subset=["temperature","event"])             #aise kr ke real time me hum subset=["Mobile NO."] likh skte hai.
```

```python
df.drop_duplicates(inplace=True)
df
```

- **7. replace() method**
    - replace method kisi data ki values ko replace kr deta hai kisi meaningfull values se.

```python
#df.replace({-9999:30})    #this approach is wrong because it will be same in every where , which where values is -9999
df.replace({'temperature':-99999},30,inplace=True)      #value=<no_default>
df.replace({'windspeed':[-99999,-88888]},10,inplace=True)
```

```python
df
```
