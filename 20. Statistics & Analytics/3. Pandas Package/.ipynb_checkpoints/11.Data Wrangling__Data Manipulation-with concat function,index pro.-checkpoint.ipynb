```python
def cal_tax(bill):
    if(bill<10):
        return bill*.05
    elif(bill>=10 and bill<20):
        return bill*.12
    else:
        return bill*.18
```

```python
cal_tax(651)
```

```python
cal_tax(50)
```

- **1. apply() method**
    - apply method total_bill wale column pe jo function humne banaya hai cal_tax ko apply kiya gya hai.

```python
df.total_bill.apply(cal_tax)
```

```python
df['tax']=df.total_bill.apply(cal_tax)
df
```

```python
df['final_bill']=df.total_bill+df.tax
df
```

```python
#update krna final_bill ko 
df['final_bill']=df.total_bill+df.tax+df.tip
df
```

- **it also includes combining multiple dataframes**
#### Combine krna--------data transformation krna
- kisi 2 dataframe ko ek sath combined krna chahte hai.-----df1+df2
- ydi humpe 2 data ho aur un dono data ka column name same ho to row wise combined krna prefered krte hai.

```python
df_cs=pd.read_csv("e:/dataset/analysis/cs.txt")
df_cs
```

```python
df_ec=pd.read_csv("e:/dataset/analysis/ec.txt")
df_ec
```

#### **concat function**
- **append**------>concat row wise
- **join(common--index)**----->concat column wise
- **merge(common--column)**------->concat column wise
- ye 2 ya 2 se jyada data ko combined krne ka kam krti hai-------but data ke columns same hone chaiye.

```python
#2 data ko combine krna
df_res=pd.concat([df_cs,df_ec],axis=0)
df_res
```

```python
df_res.loc[[0]]
```

```python
#2 data ko combine krna-----with ignore_index=True
df_res=pd.concat([df_cs,df_ec],axis=0,ignore_index=True)
df_res
```

```python
df_res.loc[[0]]
```

```python
#columns ko alphabet me convert krna--------wiht sort=True argument
df_res=pd.concat([df_cs,df_ec],axis=0,ignore_index=True,sort=True)
df_res
```

#### **reset_index() method**
- reset_index method ---strarting from 0 se start kr sake.-----default index
- extra work-------ye ek index name se column bana deta hai.

```python
df_res.reset_index()
```

```python
#drop=False jo exiting index hai use drop kr deta hai.
df_res.reset_index(drop=True)
```

- **concat function ka dusra tarika(behaviour)**

```python
#hum ek dataframe bana rahe hai jisse aage ka work kr sake
dfx=pd.DataFrame({'roll':[1,2,3],'name':['anil','bablu','cine']})
dfy=pd.DataFrame({'age':[20,22,19],'branch':['cs','ec','cs']})
```

```python
dfx
```

```python
dfy
```

- join(common--index)----->concat column wise
- **1st tarika**---concat function se join krne ka

```python
#columns index ko mearg kr dege is bari.
pd.concat([dfx,dfy],axis=1)
```

- **2nd tarika of join krna with index wise**
#### **join() method**
- join method ka disadvantage ye ek bari me sirf 2 hi data ko join kr skta hai. but concat function ek bari me kitne bhi data ko join kr skta hai.

```python
dfx.join(dfy)        #join bydefault column ko considerd krta hai.    #but iske index same hone chaiye.
```
