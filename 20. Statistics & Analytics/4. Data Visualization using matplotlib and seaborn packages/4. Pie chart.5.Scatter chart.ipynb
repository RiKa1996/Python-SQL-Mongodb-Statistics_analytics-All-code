### **Pie chart**
- pie chart hume persentage me batata hai.
- >comparision of a categorical values based on aggregate operation
- >we can easily show ratio in comparision

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### **Pie() function**

```python
df=pd.read_csv("e:/dataset/analysis/titanic.csv")
df
```

```python
val_ct=df.Pclass.value_counts()
val_ct
```

```python
#x argument and label argument-----x=val_ct.values,labels=val_ct.values------wedgeprops={'edgecolor':'k'}
plt.pie(x=val_ct.values,labels=val_ct.values,autopct="%.2f",wedgeprops={'edgecolor':'k'},startangle=90,explode=(0,.1,0))
plt.show()
```
