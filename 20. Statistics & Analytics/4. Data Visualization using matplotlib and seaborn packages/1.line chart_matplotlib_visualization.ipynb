### **1. plot() function**
- plot funciton ka kam hota hai **line chart** banana
- 2 list **,** 2 numpy array **,** 2 series **,** 2 1d array jaisa --------hum plot function me likhna padta hai.
- **plt.xticks function**------xticks function se x axis pe hum khud se marking kr skte hai.
- **plt.yticks function**------yticks function se y axis pe hum khud se marking kr skte hai.
- **plt.xlabel funciton**-----ye x axis per label lagane ke liye ye function hota hai. jaise-----Overs
- **plt.ylabel function**-----ye y axis per label lagane ke liye ye function hota hai. jaise-----runs
- **plt.text(where,what) function**------where=kha likhna hai----------what=kya likhna hai.
- **plt.title function** -----title function chart ke upper header likhne ke liye kiya jata hai.
- **plt.legend([]) function**-------legend fucntion line chart me jo line bani hui hai **uski identitiy batati hai.**
- **plt.grid fucntion**----------background ko plan se hata kr grid line me convert kr deta hai.
- **plt.savefig function**-----abhi tak jo bhi humne matplotlib se krwaya hai use hum ek figuar me save kr skte hai.

```python
over=[5,10,15,20]
runs=[50,80,130,200]
plt.plot(over,runs)
plt.xticks(range(0,22,2))
plt.yticks(range(0,250,20))
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.show()
```

```python
#over time -------line ups and down ho skte hai.
#Ie6=1*10 ki power 6 mtlb 10 ke aage 4 zero
years=[2018,2019,2020,2021,2022,2023,2024]
sales=[1000000,1500000,600000,800000,1200000,1600000,2000000]
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(years,sales,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k',)     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,
plt.xticks(range(2018,2025,1))
plt.yticks(range(500000,2500000,200000))
plt.xlabel("Years")
plt.ylabel("Sales")
plt.text(2018,1100000,"10L")                          #plt.text(where(x-axix,y-axis),what)
plt.show()
```

- **for loop laga ke----- "plt.text function"** se sabhi jagah ek hi sath value ko likh dena.

```python
#over time -------line ups and down ho skte hai.
#Ie6=1*10 ki power 6 mtlb 10 ke aage 4 zero
years=[2018,2019,2020,2021,2022,2023,2024]
sales=[1000000,1500000,600000,800000,1200000,1600000,2000000]
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(years,sales,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k',)     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,
plt.xticks(range(2018,2025,1),c='r',size=10)
plt.yticks(range(500000,2500000,200000),c='c',size=10)
plt.xlabel("Years",c="b",size=12)
plt.ylabel("Sales",c="r",size=12)
plt.title("B/S over time",size=15,c="g")
for y,s in zip(years,sales):
    plt.text(y-.2,s+50000,s,c='k',size=9)              ##plt.text(where,what)
plt.show()
```

- **Ab ke aur data ko mention karege uska bhi isi chart me line chart banayege**

```python
years=[2018,2019,2020,2021,2022,2023,2024]
sales_1=[1200000,1500000,600000,800000,1200000,1600000,2000000]
sales_2=[1000000,1300000,1600000,1800000,2200000,2600000,2900000]
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(years,sales_1,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k',)     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,
plt.plot(years,sales_2,c='b',linewidth=.8,linestyle='solid',marker="*",markerfacecolor='r',markersize=5,markeredgecolor='k',)     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,

plt.xticks(range(2018,2025,1),c='r',size=10)
plt.yticks(range(500000,3000000,200000),c='c',size=10)
plt.xlabel("Years",c="b",size=12)
plt.ylabel("Sales",c="r",size=12)
plt.title("B/S over time",size=15,c="g")
for y,s1 in zip(years,sales_1):
    plt.text(y-.2,s1+50000,s1,c='m',size=9)              ##plt.text(where,what)
for y,s2 in zip(years,sales_2):
    plt.text(y-.2,s2-90000,s2,c='b',size=9)              ##plt.text(where,what)
plt.legend(["ABC","XYZ"])         #plt.legend([company name_1,company name_2])
plt.show()
```

```python
#sbse phle data types check krte hai yhi ydi datetime data nhi hai to use phle convert krege.====to_datetime datatypes me
df["Order_Datetime"]=pd.to_datetime(df["Order Date"])
```

```python
#ab hum jo yera column banaya hai , year column ke jariye sales column se-----groupby karege.
res_df=df.groupby(by='years')[["Sales"]].sum()
res_df
```

```python
Years=res_df.index
Sales=res_df.Sales.values
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(Years,Sales,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k')     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,

plt.xlabel("Years",c="b",size=12)
plt.ylabel("Sales",c="r",size=12)
plt.title("B/S over time",size=15,c="g")
for y,s in zip(Years,Sales):
    plt.text(y-.2,s+5000,s,c='m',size=9)              ##plt.text(where,what)
plt.legend(["ABC"])         #plt.legend([company name_1,company name_2])
plt.grid(c='y')
plt.show()
```

### **line chart ki width and height ko control krna ho to**
- **figure function**
    - isse line chart ki width aur height ko control krna.

```python
Years=res_df.index
Sales=res_df.Sales.values
#khud se width aur height ko control krna
plt.figure(figsize=(10,7))      #(width,height)
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(Years,Sales,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k')     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,

plt.xticks(range(2020,2024,1))
plt.xlabel("Years",c="b",size=12)
plt.ylabel("Sales",c="r",size=12)
plt.title("B/S over time",size=15,c="g")
for y,s in zip(Years,Sales):
    plt.text(y-.2,s+5000,s,c='m',size=9)              ##plt.text(where,what)
plt.legend(["ABC"])         #plt.legend([company name_1,company name_2])
plt.grid(c='y')
plt.show()
```

```python
import pandas as pd
df=pd.read_excel("e:/dataset/Sample-Superstore.xlsx.xlsx")
df["Order_Datetime"]=pd.to_datetime(df["Order Date"])
df["years"]=df.Order_Datetime.dt.year
res_df=df.groupby(by='years')[["Sales"]].sum()

Years=res_df.index
Sales=res_df.Sales.values
#khud se width aur height ko control krna
plt.figure(figsize=(12,7))      #(width,height)
#c='b'---color ke liye, linewidth=.8, linestyle='dashed','dotted','dashdot','solid', 
plt.plot(Years,Sales,c='m',linewidth=.8,linestyle='solid',marker="o",markerfacecolor='r',markersize=5,markeredgecolor='k')     #r=red,y=yellow,g=green,b=blue,k=black,pink,m=megenta,

plt.xticks(range(2020,2024,1))
plt.xlabel("Years",c="b",size=12)
plt.ylabel("Sales",c="r",size=12)
plt.title("B/S over time",size=15,c="g")
for y,s in zip(Years,Sales):
    plt.text(y-.15,s+5000,s,c='k',size=9)              ##plt.text(where,what)
plt.legend(["ABC"])         #plt.legend([company name_1,company name_2])
plt.grid(c='y')
plt.show()
```
