## **SEABORN**
- **Matplotlib ka extension seaborn hai.**
- jb hum matplotlib lib import krte hai.-----seaborn lib autometic aa jata hai.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

#### **scatterplot() function**

```python
df=pd.read_csv("e:/dataset/analysis/restaurant.csv")
#df
```

```python
#gender wise color
sb.scatterplot(x=df.total_bill,y=df.tip,c='m',hue=df.gender)
plt.xlabel("Total_Bill",c='m')
plt.ylabel('Tip')
plt.show()            #seaborn me show krne ka function nhi chalta hai.---ise chalane ke hume matplotlib ka hi function use krna padta hai.
```

```python
#day wise color
sb.scatterplot(x=df.total_bill,y=df.tip,c='m',hue=df.day,style=df.gender)
plt.xlabel("Total_Bill",c='m')
plt.ylabel('Tip')
plt.show()            #seaborn me show krne ka function nhi chalta hai.---ise chalane ke hume matplotlib ka hi function use krna padta hai.
```

#### **countplot() function**

```python
sb.countplot(x=df.day)
plt.show()
```

```python
sb.countplot(x=df.day,hue=df.time)
plt.show()
```

```python
#day ko hum apna order de skte hai ---------order argument se 
sb.countplot(x=df.day,order=['Thur','Fri','Sat','Sun'])
plt.show()
```

```python
#day ko hum apna order de skte hai ---------order argument se 
sb.countplot(y=df.day,order=['Thur','Fri','Sat','Sun'])
plt.show()
```

```python
sb.lineplot(x=df.day,y=df.gender)
plt.show()
```

```python
sb.histplot(x=df.total_bill)
plt.show()
```

```python
sb.pairplot(hue=df['day','total_bil'])
```
