#### **boxplot() function**

```python
x=[10,11,12,10.5,11.5,12.50,510,11,100,-100]
sb.boxplot(x)
```

```python
import numpy as np
x=np.random.randint(1,50,100)

sb.boxplot(x)
plt.show()
```

```python
import numpy as np
x=np.random.randint(1,50,100)
x[10]=150          #high outlier
x[12]=100         #low outlier
x[15]=-50          #hig outlier
x[20]=-150
sb.boxplot(x)
plt.show()
```

- **orient argument**

```python
import numpy as np
x=np.random.randint(1,50,100)
x[10]=150          #high outlier
x[12]=100         #low outlier
x[30]=-50          #hig outlier

#sb.boxplot(x)
sb.boxplot(x,orient="h")
plt.show()
```

```python
import numpy as np
x=np.random.randint(1,50,100)
y=np.random.randint(1,50,100)

x[10]=150          #high outlier
x[12]=100         #low outlier
x[30]=-50          #hig outlier

sb.boxplot([x,y])
plt.show()
```

### **11. Subplot**

### **subplot() function**
- ek figure pe multiple chart ko mention krna.
- ek figure me ydi 2x2=4 figure,3

```python
plt.figure(figsize=(7,7))
plt.subplot(2,2,1)
sb.countplot(x=df.day)

plt.subplot(2,2,2)
sb.scatterplot(x=df.total_bill,y=df.tip)

plt.subplot(2,2,4)
x=np.random.randint(1,50,100)
x[10]=150          #high outlier
x[12]=100         #low outlier
x[30]=-50          #hig outlier
sb.boxplot([x,y])

plt.subplot(2,2,3)
sb.heatmap(df.corr())
plt.show()
```
