### **Bar Chart**
- bar chart pe x-axis per categorical data aata hai.
- Kisi catogorical data ka, column ka aggregate operation  krna
- >to compare result of aggregate operation for a categorical column.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("e:/dataset/analysis/restaurant.csv")
df.day.value_counts()
```

```python
res_df=df.groupby(by="day")[["total_bill"]].agg(['count','sum','mean','median'])     #aggregate
res_df
```

```python
res_df["total_bill"]["count"].values
```

### **bar() function**

```python
bar=plt.bar(res_df.index,res_df["total_bill"]["count"].values,edgecolor='k',color=["r",'g','m','y'],)
bar
```

```python
bar=plt.bar(res_df.index,res_df["total_bill"]["mean"].values,edgecolor='k',color=["r",'g','m','y'],)
bar
```

```python
bar=plt.bar(res_df.index,res_df["total_bill"]["median"].values,edgecolor='k',color=["r",'g','m','y'],)
bar
```
