## **Web Scraping**
    - **statistic data**---ye wo data hote hum jo publicly hoti hai.jaise ----who ki web site ya covid,because ispe hum koi apna account nhi open krne wale hai.
    - **dynamic data**-----like ,facebook,instagram,irctc aur bhi aisi website jispe hum bhi jhud skte hai.
- web scraping ka work hota hai kisi web site se data ko collect krna.
- jo web site se data ko collect kiya,---uske ander se meaning full data, ya pars krna.
- **sir ki defination**
- It is process of collecting data from web and parsing it accordingly
- Some libraries ya module of collecting data
    - **requests module ya library**
        - **requests module** provides facility to collect data and **bs4 module** porvides facility to parse data. 
        - requests module ka kam hota hai **data ko collect krna**
    - **bs4 module ya library**
        - jo data **requests module** se aa gaya hai use meaningfull banane ke liye **bs4 module** ka use krte hai.
        

```python
import requests
import bs4
```

#### **1. get() function**

```python
#https://www.mygov.in/corona-data/covid19-statewise-status/------------this is web site of covid
con=requests.get("https://www.mygov.in/corona-data/covid19-statewise-status/")
con
```

```python
data=con.text
print(data)
```

#### **A. BeautifulSoup**-----bs4 me CLASS hai. 

```python
#bs4 module ka use jo data humne collect kiya use meaningfull banane ke liye kiya jata hai.
soup=bs4.BeautifulSoup(data)
```

#### **a. find method**

```python
soup.find("title")
```
